# Discrete FDR tutorial

Dependencies

```python
import pandas as pd
from biom import load_table
from dsfdr import dsfdr
```

Load data

```python
data = load_table('deblur-feature-table.biom').to_dataframe()
data.shape
```

```python
data.head()
```

Load metadata

```python
meta = pd.read_csv('metadata_rare2k.txt', sep='\t', index_col=0).loc[data.columns]
meta.shape
```

```python
meta.head()
```

Convert categories into incremental integers

```python
labels = meta['exposure_type'].astype('category').cat.codes
```

Run dsFDR

```python
result = dsfdr.dsfdr(data.values, labels, transform_type='rank', method='meandiff',
                     alpha=0.01, numperm=10000, fdr_method='dsfdr')
```

Display result

```python
df = pd.DataFrame(result).transpose()
df.columns = ['rejected', 'test_statistic', 'pvalue']
df.index = data.index
df.head(10)
```

Output result

```python
df.to_csv('output.tsv', sep='\t')
```
